```python
dataset = []
folder_paths = [
    "drive/My Drive/data/fruits-360/fruits-360-original-size/fruits-360-original-size/Training/apple_6",
    "drive/My Drive/data/fruits-360/fruits-360-original-size/fruits-360-original-size/Training/apple_braeburn_1",
    "drive/My Drive/data/fruits-360/fruits-360-original-size/fruits-360-original-size/Training/apple_crimson_snow_1",
    "drive/My Drive/data/fruits-360/fruits-360-original-size/fruits-360-original-size/Training/apple_golden_1",
    "drive/My Drive/data/fruits-360/fruits-360-original-size/fruits-360-original-size/Training/apple_golden_2",
    "drive/My Drive/data/fruits-360/fruits-360-original-size/fruits-360-original-size/Training/apple_golden_3"
]
```

```python
import os
import cv2
for i in folder_paths:
  folder_name = os.path.basename(i)

  for file_name in os.listdir(i):
    image_path = os.path.join(i, file_name)

    if os.path.isfile(image_path):
      image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
      if image is not None:
        resized_image = cv2.resize(image, (250,250))
        flattened_image = resized_image.flatten().tolist()
        dataset.append(flattened_image + [folder_name])
```

```python
import pandas as pd
df = pd.DataFrame(dataset, columns = [f"pixel {i+1}" for i in range (250*250)] + ["label"])
df
```

```python
x = df.drop("label", axis =1)
x = x/255
y = df["label"]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

y_encoded = encoder.fit_transform(y)
```

```python
y_series = pd.Series(y_encoded, name = "Target")

df = pd.concat([x, y_series], axis = 1)
df
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
np.random.seed(32)
x_train, x_test, y_train, y_test = train_test_split(x, y_series, test_size = 0.2)
```

```python
from keras.models import Sequential
model = Sequential()
```

```python
from keras.layers import Dense, Flatten
```

```python

model.add(Dense(128, activation='relu', input_shape=(250*250,)))


model.add(Dense(64, activation='relu'))


model.add(Dense(len(encoder.classes_), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
from keras.utils import to_categorical
y_train_encoded  = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)
```

```python
history = model.fit(x_train, y_train_encoded, batch_size = 32, epochs = 10, validation_split =0.2)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred_encoded = model.predict(x_test)

y_pred = np.argmax(y_pred_encoded, axis =1)

accuracy = accuracy_score(y_test, y_pred)
print("accuracy: ", accuracy)

print(classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)
print("confusion matrix")
print(conf_matrix)
```
